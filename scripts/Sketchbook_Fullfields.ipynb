```python
from plasmeep.lib import Plasmeep as pm
from meep import mpb
import scipy.special as sp
import meep as mp
import numpy as np
import csv
import matplotlib.pyplot as plt
import os
```

```python
output = os.getcwd()+'/../outputs/plots/'
savefile = output+'bands.csv'

## Parameters for simulation
res = 128
nx = 14
ny = 12
dpml = 1
a = 0.028
B = 0.05
sim = pm(a, res, dpml, nx, ny, B=np.array([0,0,-B]))

fs = 3.85*10**9
fp = 8*10**9
gamma = 1*10**6
F = np.array([fs,fp,gamma])
F = sim.Nondimensionalize_Freq(F)

eps = 1.0 #Background Permittivity

## Build Geometry
a1 = np.array([0,1,0])
a2 = np.array([np.sqrt(3)/2,1./2,0])
a_basis = np.array([a1,a2])
G_w = 0.078/a

## PEC enclosure
sim.Add_Prism(np.array([(3+1./np.sqrt(3))*a1, (3.5+1./np.sqrt(3))*a1,\
             (3.5+1./np.sqrt(3))*a1-4.5*a2, (3+1./np.sqrt(3))*a1-4*a2]), PEC = True)
sim.Add_Prism(np.array([(3+1./np.sqrt(3))*a1-4*a2, (3.5+1./np.sqrt(3))*a1-4.5*a2,\
                        [-3*np.sqrt(3)/2-1./2-np.sqrt(3)/4, G_w/2,0],\
                        [-3*np.sqrt(3)/2-1./2, G_w/2,0]]), PEC = True)
sim.Add_Prism(np.array([(3+1./np.sqrt(3))*a1, (3.5+1./np.sqrt(3))*a1,\
               (3.5+1./np.sqrt(3))*a2, (3+1./np.sqrt(3))*a2]), PEC = True)
sim.Add_Prism(np.array([(3.5+1./np.sqrt(3))*a2, (3+1./np.sqrt(3))*a2,\
                        [3*np.sqrt(3)/2+1./2+np.sqrt(3)/4, G_w/2,0],\
                        [3*np.sqrt(3)/2+1./2, G_w/2,0]]), PEC = True)
sim.Add_Prism(np.array([-(3+1./np.sqrt(3))*a1, -(3.5+1./np.sqrt(3))*a1,\
               -(3.5+1./np.sqrt(3))*a1+4.5*a2, -(3+1./np.sqrt(3))*a1+4*a2]), PEC = True)
sim.Add_Prism(np.array([-(3+1./np.sqrt(3))*a1+4*a2, -(3.5+1./np.sqrt(3))*a1+4.5*a2,\
                        [3*np.sqrt(3)/2+1./2+np.sqrt(3)/4, -G_w/2,0],\
                        [3*np.sqrt(3)/2+1./2, -G_w/2,0]]), PEC = True)
sim.Add_Prism(np.array([-(3+1./np.sqrt(3))*a1, -(3.5+1./np.sqrt(3))*a1,\
               -(3.5+1./np.sqrt(3))*a2, -(3+1./np.sqrt(3))*a2]), PEC = True)
sim.Add_Prism(np.array([-(3.5+1./np.sqrt(3))*a2, -(3+1./np.sqrt(3))*a2,\
                        [-3*np.sqrt(3)/2-1./2-np.sqrt(3)/4, -G_w/2,0],\
                        [-3*np.sqrt(3)/2-1./2, -G_w/2,0]]), PEC = True)
## Entrance/Exits
sim.Add_Block((-7,-G_w/2-np.sqrt(3)/4,0), (7-3*np.sqrt(3)/2-1./2,np.sqrt(3)/4,mp.inf), PEC = True)
sim.Add_Block((-7,G_w/2,0), (7-3*np.sqrt(3)/2-1./2,np.sqrt(3)/4,mp.inf), PEC = True)
sim.Add_Block((3*np.sqrt(3)/2+1./2,-G_w/2-np.sqrt(3)/4,0), (7-3*np.sqrt(3)/2-1./2,np.sqrt(3)/4,mp.inf), PEC = True)
sim.Add_Block((3*np.sqrt(3)/2+1./2,G_w/2,0), (7-3*np.sqrt(3)/2-1./2,np.sqrt(3)/4,mp.inf), PEC = True)
## Crystal
sim.Rod_Array_Hexagon(np.array([0,0,0]), 4, r = 0.005/a, a = 1, eps = 1,\
                          a_basis = np.array([[0,1,0], [np.sqrt(3)/2,1./2,0]]),\
                          bulbs = False, wp = F[1], gamma = F[2],\
                          axis = np.array([0,0,1]), height = mp.inf)

## Add Source and Simulate
sim.Add_Cont_Source(0.42, [0,0,1], [-5,0,0], [0,G_w,0], E = False)

sim.Viz_Domain(output, F[0])
sim.Sim_And_Plot(1.0, 'H', 100, output, 'hz', out_prefix = 'highres_B+', remove = False)
```
